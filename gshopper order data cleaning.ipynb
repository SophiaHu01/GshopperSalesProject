```python
loyal_customers = df[['Account Name','Total Orders','Total Order Quantity','Total Payment', '1st Interaction Channel']].sort_values(by='Total Order Quantity', ascending = False)
```

```python
# drop the duplicate rows
loyal_customers = loyal_customers.drop_duplicates(subset=['Account Name'])
loyal_customers
```

```python
#define the level of the customer based on their purchasing history
def loyalcustomer(purchase):
    if purchase >= 6:
        level = 'High'
    elif 3<= purchase < 6:
        level = 'Medium'
    elif purchase == 1:
        level = 'New Customer'
    else:
        level = 'Low'
    return level
```

```python
loyal_customers['Customer Loyalty'] = loyal_customers['Total Order Quantity'].apply(lambda x : loyalcustomer(x))
```

```python
loyal_customers
```

```python
loyal_customers.to_excel('CRM Data.xlsx', index=False)
```
